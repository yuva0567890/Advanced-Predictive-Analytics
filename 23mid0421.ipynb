```python
X1 = df1.drop("MedHouseVal", axis=1)
y1 = df1["MedHouseVal"]
```

```python
##CREATING,TRAINING,TESTING ,PREDICTING,EVALUATING THE MODELS AND ALSO SAVING THE RESULTS

print("(i) TRAIN/TEST SPLIT")
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42
)
```

```python
print("FEATURE SCALING")
from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler()

X1_train = scaler1.fit_transform(X1_train)
X1_test = scaler1.transform(X1_test)
```

```python
print("CHECKING SHAPES")
print("Training Data:", X1_train.shape)
print("Testing Data :", X1_test.shape)
```

```python
#SIMILARLY FOR DATASET 2
print(df2.columns)
```

```python
X2 = df2.drop("Y house price of unit area", axis=1)
y2 = df2["Y house price of unit area"]
```

```python
from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()

X2_train = scaler2.fit_transform(X2_train)
X2_test = scaler2.transform(X2_test)
```

```python
print("Training Data:", X2_train.shape)
print("Testing Data :", X2_test.shape)
```

```python
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import numpy as np
```

```python
results = {}
```

```python
#MODELS 1--LINEAR REGRESSION
print("Models")
print("Linear Regression")
lr_model = LinearRegression()
lr_model.fit(X1_train, y1_train)
y1_pred_lr = lr_model.predict(X1_test)
mae = mean_absolute_error(y1_test, y1_pred_lr)
mse = mean_squared_error(y1_test, y1_pred_lr)
rmse = np.sqrt(mse)
r2 = r2_score(y1_test, y1_pred_lr)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
results["Linear Regression"] = [mae, mse, rmse, r2]
```

```python
#MODELS 2--POLYNOMIAL REGRESSION
print("Polynomial Regression")
poly_model = Pipeline([
    ("poly", PolynomialFeatures(degree=2)),
    ("linear", LinearRegression())
])

poly_model.fit(X1_train, y1_train)
```

```python
y1_pred_poly = poly_model.predict(X1_test)
mae = mean_absolute_error(y1_test, y1_pred_poly)
mse = mean_squared_error(y1_test, y1_pred_poly)
rmse = np.sqrt(mse)
r2 = r2_score(y1_test, y1_pred_poly)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
results["Polynomial Regression"] = [mae, mse, rmse, r2]
```

```python
#MODELS 3--ELASTIC NET REGRESSION
print("Elastic Net")
elastic_model = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
elastic_model.fit(X1_train, y1_train)
```

```python
y1_pred_elastic = elastic_model.predict(X1_test)
mae = mean_absolute_error(y1_test, y1_pred_elastic)
mse = mean_squared_error(y1_test, y1_pred_elastic)
rmse = np.sqrt(mse)
r2 = r2_score(y1_test, y1_pred_elastic)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
results["Elastic Net"] = [mae, mse, rmse, r2]
```

```python
#MODELS 4--RANDOM FOREST REGRESSION
print("Random Forest Regressor")
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X1_train, y1_train)
```

```python
y1_pred_rf = rf_model.predict(X1_test)
mae = mean_absolute_error(y1_test, y1_pred_rf)
mse = mean_squared_error(y1_test, y1_pred_rf)
rmse = np.sqrt(mse)
r2 = r2_score(y1_test, y1_pred_rf)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
results["Random Forest"] = [mae, mse, rmse, r2]
```

```python
#MODELS 5-- GRADIENT BOOSTING REGRESSION
print("Gradient Boosting regressor")
gb_model = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

gb_model.fit(X1_train, y1_train)
```

```python
y1_pred_gb = gb_model.predict(X1_test)
mae = mean_absolute_error(y1_test, y1_pred_gb)
mse = mean_squared_error(y1_test, y1_pred_gb)
rmse = np.sqrt(mse)
r2 = r2_score(y1_test, y1_pred_gb)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
results["Gradient Boosting"] = [mae, mse, rmse, r2]
```

```python
#COMPARISON
print("Comparison of Tables")
comparison = pd.DataFrame(
    results,
    index=["MAE", "MSE", "RMSE", "R2 Score"]
).T

comparison
```

```python
#COMPARING MODELS
comparison = pd.DataFrame(
    results,
    index=["MAE", "MSE", "RMSE", "R2 Score"]
).T

comparison
```

```python
#FINDING BEST MODELS
best_model = comparison["R2 Score"].idxmax()

print("Best Model :", best_model)
print("Highest R2 Score :", comparison.loc[best_model, "R2 Score"])
```

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing(as_frame=True)
df1 = housing.frame
```

```python
X1 = df1.drop("MedHouseVal", axis=1)
y1 = df1["MedHouseVal"]
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1,
    y1,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler()

X1_train = scaler1.fit_transform(X1_train)
X1_test = scaler1.transform(X1_test)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X1_train, y1_train)

y1_pred_lr = lr_model.predict(X1_test)
```

```python
print(y1_test.shape)
print(y1_pred_lr.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.scatter(y1_test, y1_pred_lr)

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Linear Regression")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(y1_test.values[:100], label="Actual")
plt.plot(y1_pred_lr[:100], label="Predicted")

plt.title("Actual vs Predicted (Linear Regression)")
plt.xlabel("Sample")
plt.ylabel("House Price")

plt.legend()
plt.grid(True)

plt.show()
```
